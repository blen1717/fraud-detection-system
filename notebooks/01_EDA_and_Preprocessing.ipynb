# EDA for Fraud Detection (Task 1)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
fraud = pd.read_csv('/content/data/raw/Fraud_Data.csv')
credit = pd.read_csv('/content/data/raw/creditcard.csv')
print('E-commerce fraud rate:', fraud['class'].mean())
print('Credit fraud rate:', credit['Class'].mean())
fig, axes = plt.subplots(1,2, figsize=(10,4))
fraud['class'].value_counts(normalize=True).plot(kind='bar', ax=axes[0], title='E-commerce')
credit['Class'].value_counts(normalize=True).plot(kind='bar', ax=axes[1], title='Credit Card')
plt.show()
```
